# Text Processing

```python
import numpy as np  
import pandas as pd
import random
```

```python
with open('greatgatsby.txt', 'r') as file:
    data = file.read().replace('\n', ' ')
data
```

Read file as string code source: https://www.tutorialkart.com/python/python-read-file-as-string/

## Cleaning the Code

removing capitalization: 

```python
data = data.lower()
```

removing punctuation:

```python
punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~“”'''
no_punct = ''
for char in data:
    if char not in punctuations:
        no_punct = no_punct + char
print(no_punct)
```

punctuation code source: https://www.programiz.com/python-programming/examples/remove-punctuation

## Splitting Data

```python
book_list = no_punct.split(' ')

#removing the empty strings
without_empty_strings = []
for string in book_list:
    if (string != ""):
        without_empty_strings.append(string)
        
without_empty_strings
```

## Getting Unique Words

```python
def unique(without_empty_strings):
    x = np.array(without_empty_strings)
    return np.unique(x)
unique = unique(without_empty_strings)
unique
```

for unique code: https://www.geeksforgeeks.org/python-get-unique-values-list/

## Create dictonary

```python
text_dict = {word: [] for word in unique}

for i in range (len(without_empty_strings)-2):
    text_dict[without_empty_strings[i]].append(without_empty_strings[i +1])
    
text_dict
```
