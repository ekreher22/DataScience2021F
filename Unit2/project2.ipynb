# What Major Should You Go Into?

# If you want to go into a female dominated field, here are your best options

```python
percentages = (majors["Women"]/majors["Total"]) * 100
percentages = pd.Series(percentages)
majors['percentages'] = percentages
percentages = majors.loc[(percentages > 90), ['Major', 'percentages']]
percentages = percentages.sort_values('percentages', ascending = False)
top_percentages = percentages.head(11)
print('Here are the top 10 women dominated fields:')
top_percentages
```

To find the female dominated fields I made a series that includes the number of women divided by number of men times 100 for each major. To get all fields that were female dominated, I only took the ones where the percentage is over 100, then printed the top 10 for cleanness (I used method chaining to get the top 10 values).

# Would you like to have a good chance of being employed? Maybe dont go into this major because it is most likely to leave you unemployed! 

```python
max_unemployment = majors.Unemployment_rate.max()
unemployment = majors.loc[(majors.Unemployment_rate == max_unemployment), ['Major', 'Unemployment_rate']]
unemployment
```

To find the job with the least employment rate I used the unemployment_rate column and I summarized the data to find the max value by using max(), then returned the Major name and the Unemployment rate by splicing. 

# Du you want to make alot of money? This type of major makes the most money on average. 

```python
major_sum_of_income = majors.groupby('Major_category').Median.sum()
major_size = majors.groupby('Major_category').size()
average = major_sum_of_income / major_size
average.loc[average == average.max()]
```

When finding what type of major makes the most money, I grouped data by using groupby to create an object with all major types, and I used method chaining on the groupby to find the sum of the Median for each type of major. I divided the sum of the income and the size of the major to get the average and then returned the max value to reveal that engineering pays the most.

# Which engineering major pays the most?

```python
max_value = majors.Median.max()
most_payed_major = majors.loc[(majors.Major_category == 'Engineering') & (majors.Median == max_value),['Major', 'Median'] ] 
most_payed_major
```

To find which engineering major makes the most, I used method chaining to define the max median. I then printed the only major that was labeled as engineering & its median was equivilent to the max median. I spliced the data to show only the Major and Median of the result.

# Don't like STEM? What non-STEM major with pay the highest?

```python
non_stem = majors.loc[(majors.Major_category != 'Engineering') & (majors.Major_category != 'Biology & Life Science') & (majors.Major_category != 'Computers & Mathematics') & (majors.Major_category != 'Health') & (majors.Major_category != 'Physical Sciences')]
max_value = non_stem.Median.max()
non_stem.loc[(non_stem.Median == max_value), ['Major', 'Major_category', 'Median']]
```

I made a variable that only returned majors that are not any stem related topic, then for the max median of those majors using method chaining. I then prined the major that had the same Median as the maximum median and only showed the Major, Major_category, and Median by splicing.

# What major is best for you if you want a stable full year round job?

```python
num_of_ppl_full_year = (majors["Total"]/majors["Full_time_year_round"])
majors['num_of_ppl_full_year'] = num_of_ppl_full_year
num_of_ppl_full_year = majors.loc[(num_of_ppl_full_year > 1), ['Major', 'num_of_ppl_full_year']]
num_of_ppl_full_year = num_of_ppl_full_year.sort_values('num_of_ppl_full_year',ascending = False)
num_of_ppl_full_year = num_of_ppl_full_year.head(11)
num_of_ppl_full_year
```

To find the majors with the highest number of people in a stable full year job relative to the total size of the major, I divided the Total column by the Full_time_year_round column. I added the quotients to the dataframe as a column and then printed the top 10 majors with the highest rate of people in a stable year round job.
