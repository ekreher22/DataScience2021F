```python
import pandas as pd
import seaborn as sns
```

```python
crimes = pd.read_csv('crime.csv')
crimes
```

```python
# do bar graph for how many crimes happened each month and compare them 
```

```python
crimes = crimes.sort_values('MONTH', ascending = True)
month1 = crimes.loc[(crimes.MONTH == 1), ['OFFENSE_CODE_GROUP']]
month1 = month1.count()
month2 = crimes.loc[(crimes.MONTH == 2), ['OFFENSE_CODE_GROUP']]
month2 = month2.count()
months = [month1, month2]
months
```

```python
sns.barplot(x = 'MONTH' , y = crimes.index, data = crimes)
```
