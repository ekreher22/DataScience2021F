```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
people = pd.read_csv('people.csv')
people
```

```python
sns.lmplot(x = 'Num of Friends' , y = 'GPA' , data = people)
plt.show()
```

```python
reviews = pd.read_csv('winemag-data_first150k.csv') 
reviews
```

```python
best_wines = reviews.loc[reviews.points == 100] 
best_wines
```

```python
num_of_best = best_wines.value_counts('country')
sns.barplot(x = num_of_best.index, y = num_of_best)
plt.grid(axis = 'y')
plt.show()
```
